#### Подготовка выборки для обучения, валидации и теста

Разобьем данные на выборки. Для исследования целевым признаком является параметр `Exited`, все остальные - это признаки, по которым мы будем учить модель определять покинет клиент банк или нет.

Так как мы знаем о перекосе в целевом признаке, постараемся сохранить такое же соотношение и при обучении и тестировании. Для этого воспользуемся параметром stratify по целевому признаку в train_test_split.

```python
data['Exited'].value_counts() #рассмотрим количественное распределение в целевом признаке
```

```python
features = data.drop('Exited', axis=1) # извлечем основные признаки 
target = data['Exited'] # извлечем целевой признак

features_train, features_test1, target_train, target_test1 = train_test_split(
    features, target, test_size=0.4, random_state=12345, stratify=target) # отделим 40% данных для валидационной и тестовой выборки
features_test, features_valid, target_test, target_valid =  train_test_split(
    features_test1, target_test1, test_size=0.5, random_state=12345) # поделим данные пополам для тестовой и валидационной выборки из обучающей части
#далее посмотрим, какие выборки получились по основным признакам
features_train.info()
print('\n')
features_test.info()
print('\n')
features_valid.info()
#и какое количественное соотношение получилось в целевом признаке.
print('\n---Распределение в target_train---\n')
print(target_train.value_counts())
print('\n---Распределение в target_test---\n')
print(target_test.value_counts())
print('\n---Распределение в target_valid---\n')
print(target_valid.value_counts())
```

Данные получилось разбить как и задумывалось - 60% на обучение и по 20% на тест и валидационную выборку.

#### Подготовка upsampled выборки

На данном этапе, зная о специфике данных, сразу подготовим upsampled данные.

PS. делаю это здесь, т.к после OHE мне это не удалось сделать.

```python
def upsample(features, target, repeat):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    features_upsampled = pd.concat([features_zeros] + [features_ones] * repeat)
    target_upsampled = pd.concat([target_zeros] + [target_ones] * repeat)
    
    features_upsampled, target_upsampled = shuffle(
        features_upsampled, target_upsampled, random_state=12345)
    
    return features_upsampled, target_upsampled

features_upsampled, target_upsampled = upsample(features_train, target_train, 4)

print(features_upsampled.shape)
print(target_upsampled.shape)
print('\n---Распределение в target_upsampled---\n')
print(target_upsampled.value_counts())
```

#### Применение OHE

В датасете присутствуют категориальные признаки, которые для исследования необходимо перекодировать в количественные. Для этого воспользуемся OHE.

```python
transformer = make_column_transformer(
    (OneHotEncoder(), ['Gender', 'Geography']),
    remainder='passthrough')
transformer.fit(features_train)
transformed = transformer.transform(features_train) #преобразуем сначала обучающую выборку
features_train = pd.DataFrame(transformed, columns=transformer.get_feature_names())
```

```python
features_train.head()
```

```python
transformed = transformer.transform(features_valid) #затем валидационную выборку
features_valid = pd.DataFrame(transformed, columns=transformer.get_feature_names())
features_valid.head()
```

```python
transformed = transformer.transform(features_test) #и наконец тестовую
features_test = pd.DataFrame(transformed, columns=transformer.get_feature_names())
features_test.head()
```

#### Подбор гиперпараметров моделей

Возьмем три основных модели обучения, подходящие для задач классификации, и рассмотрим на обучающей выборке, какая из них отработает лучше всего и с какими параметрами.

Для выбора оптимальных гиперпараметров воспользуемся GridSearchCV. Далее выберем самую эффективную модель и на ней поработаем с дисбалансом данных.

##### Решающее дерево

```python
model_dtc = DecisionTreeClassifier(random_state=12345)

model_dtc.get_params().keys() #напомним основные параметры
```

```python
param_grid = { 
    'max_features': ['auto', 'sqrt', 'log2'],
    'max_depth': range(1, 10, 1),
    'criterion': ['gini', 'entropy']
}
scoring = ['f1', 'roc_auc'] #отберем лучшие параметры на основе метрик
```

```python
CV_model_dtc = GridSearchCV(estimator=model_dtc, param_grid=param_grid, cv=5, refit='f1', scoring=scoring)
CV_model_dtc.fit(features_train, target_train)
```

```python
best_dtc = CV_model_dtc.best_estimator_ #сохраним лучшую модель
print(CV_model_dtc.best_params_) #выведем лучшие гиперпараметры
```

```python
predicted_valid = best_dtc.predict(features_valid)

print("F1 на валидационной выборке:", f1_score(target_valid, predicted_valid))
```

```python
metrics_columns = [f'mean_test_{x}' for x in scoring]
final_metrics = pd.DataFrame(CV_model_dtc.cv_results_)[metrics_columns].iloc[CV_model_dtc.best_index_]
print(final_metrics) #выведем метрику модели
```

## Борьба с дисбалансом

#### Применение class_weight

Попробуем поработать с дисбалансом выборки на примере этой модели и улучшим ее показатели.

Для начала добавим в гиперпараметры class_weight и снова воспользуемся GridSearchCV.

```python
model = RandomForestClassifier(random_state=12345)

param_grid = {
    'n_estimators': [100], #возьмем зафиксированное значение, чтобы не перегружать процесс подбора параметров.
#Опытным путем было определено это значение.
    'max_features': ['auto', 'sqrt', 'log2'],
    'max_depth': range(1, 20, 1),
    'criterion': ['gini', 'entropy'],
    'class_weight': ['balanced']
}
scoring = ['f1', 'roc_auc']
```

```python
CV_model = GridSearchCV(estimator=model, param_grid=param_grid, cv=5, refit='f1', scoring=scoring, n_jobs=-1)
CV_model.fit(features_train, target_train)
```

```python
best_model = CV_model.best_estimator_ #сохраним лучшую модель
print(CV_model.best_params_) #выведем лучшие гиперпараметры
```

```python
predicted_valid = best_model.predict(features_valid)

print("F1 на валидационной выборке:", f1_score(target_valid, predicted_valid))
```

```python
metrics_columns = [f'mean_test_{x}' for x in scoring]
final_metrics = pd.DataFrame(CV_model.cv_results_)[metrics_columns].iloc[CV_model.best_index_]
print(final_metrics) #выведем метрику модели
```

Сравним с метрикой логистической регрессии после добавления дополнительного гиперпараметра.

```python
CV_model_lg = GridSearchCV(estimator=model_lg, param_grid=param_grid, cv=5, refit='f1', scoring=scoring)
CV_model_lg.fit(features_train_scaled, target_train)
```

***Общий вывод:***
Здесь мы видим как сильно улучшились результаты работы логистической регрессии. Но лес все же впереди. На нем отработаем дальнейшие шаги.

#### Upsample выборки

Попробуем улучшим результат и переобучить модель леса на upsampled обучающей выборки, которую мы подготовили с помощью функции upsample. Подберем новые гиперпараметры с помощью цикла:

```python
best_model = None
best_result = 0
for est in range(10, 100, 10):
    for depth in range (1, 20, 1):
        model = RandomForestClassifier(random_state=12345, n_estimators=est, max_depth = depth, criterion='entropy')
        model.fit(features_upsampled, target_upsampled)
        predictions = model.predict(features_valid)
        result = f1_score(target_valid, predictions)
        if result > best_result:
            best_model = model
            best_result = result
            best_est = est
            best_depth = depth
print("F1 наилучшей модели на валидационной выборке:", best_result, '\nКоличество деревьев:', best_est, '\nГлубина:', best_depth)
```

Показатели незначительно улучшились.

#### Общий вывод

Результаты удалось улучшить благодаря class_weight=balanced, который придадает объектам редкого класса больший вес, и увеличению присутствия редкого класса в выборке.

Самое время проверить модель теперь и на тестовой выборке.

## Тестирование модели

```python
predicted_test = best_model.predict(features_test)
probabilities_test = best_model.predict_proba(features_test)
probabilities_one_test = probabilities_test[:, 1]

print("F1:", f1_score(target_test, predicted_test))
```

Результат нас устраивает, все получилось.

#### ROC-кривая
